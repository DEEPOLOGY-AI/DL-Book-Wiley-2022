```python
import torch
import torch
import decimal
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
plt.style.use("seaborn-bright")
def float_range(start, stop, step):
  while start < stop:
    yield float(start)
    start += decimal.Decimal(step)
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)

input= np.linspace(-6, 6)
output = binaryStep(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Binary Step')
plt.xlim([-6, 6])
plt.ylim([-1, 1])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()  
```

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x

x= list(float_range(-6, 6, '0.1'))
input= torch.tensor(x)
output = linear(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Linear')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()  
```

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
x = np.linspace(-6, 6)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
def elu(arr):
    lst=[]
    for i in arr:
        if i<0:
            lst.append(0.5*(np.exp(i) -1))
        else:
            lst.append(i)
    return lst
input= np.linspace(-6, 6)
output = elu(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('ELU')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.axis('tight')
plt.savefig('ELU.png')
plt.show()
```

```python
# HARDSHRINK
def hardshrink(arr):
    lst=[]
    for i in arr:
        if i>0.5:
            lst.append(i)
        elif i<-0.5:
            lst.append(i)
        else:
            lst.append(0)
    return lst
input= np.linspace(-6, 6)
output = hardshrink(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('HARDSHRINK')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.axis('tight')
plt.savefig('HARDSHRINK.png')
plt.show()
```

```python
# HARDSIGMOID
def hardSigmoid(arr):
    lst=[]
    for i in arr:
        if i>=3:
            lst.append(1)
        elif i<=-3:
            lst.append(0)
        else:
            lst.append(i/6+0.5)
    return lst
input= np.linspace(-6, 6)
output = hardSigmoid(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('HARDSIGMOID')
plt.xlim([-6, 6])
plt.axis('tight')
plt.savefig('HARDSIGMOID.png')
plt.show()
```

```python
# HARDTANH
def hardTanh(arr):
    lst=[]
    for i in arr:
        if i>1:
            lst.append(1)
        elif i<-1:
            lst.append(-1)
        else:
            lst.append(i)
    return lst
input= np.linspace(-6, 6)
output = hardTanh(input)

plt.plot(input,output, label='train')
plt.grid(True)
plt.title('HARDTANH')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('HARDTANH.png')
plt.show()
```

```python
# HARDSWISH
def hardSwish(arr):
    lst=[]
    for i in arr:
        if i>=3:
            lst.append(i)
        elif i<=-3:
            lst.append(0)
        else:
            lst.append(i*(i+3)/6)
    return lst
input= np.linspace(-6, 6)
output = hardSwish(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('HARDSWISH')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('HARDSWISH.png')
plt.show()
```

```python
# RELU
#  https://arxiv.org/abs/1603.05201
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1

input= np.linspace(-6, 6)
output = RELU(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('RELU')
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('RELU.png')
plt.show()
```

```python
# RELU6
def RELU6(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        elif i>=0 and i<=6:
            x1.append(i)
        else:
            x1.append(6)

    return x1
input= np.linspace(-8, 8)
output = RELU6(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('ReLU6')
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('ReLU6.png')
plt.show()
```

```python
# RReLU
def pRELU(x,alpha):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(alpha*i)
        else:
            x1.append(i)

    return x1

input= np.linspace(-6, 6)
output = pRELU(input,0.25)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('RReLU')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('RReLU.png')
plt.show()
```

```python
# SELU
def selu(arr):
    lst=[]
    for i in arr:
        if i<0:
            lst.append(1.6732632*(np.exp(i) -1))
        else:
            lst.append(1.05070098*i)
    return lst
input= np.linspace(-6, 6)
output = selu(input)

plt.plot(input,output, label='train')
plt.grid(True)
plt.title('SELU')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('SELU.png')
plt.show()
```

```python
# CELU
def celu(arr,alpha):
    lst=[]
    for i in arr:
        if i<0:
            lst.append(alpha*(np.exp(i/alpha)-1))
        else:
            lst.append(1.05070098*i)
    return lst
input= np.linspace(-6, 6)
output = celu(input,1)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('CELU')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('CELU.png')
plt.show()
```

```python
# GELU
def gelu(x):
    cdf = 0.5 * (1.0 + tf.math.erf(x / tf.math.sqrt(2.0)))
    return x * cdf
input= np.linspace(-6, 6)
output = gelu(input)

plt.plot(input,output, label='train')
plt.grid(True)
plt.title('GELU')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('GELU.png')
plt.show()
```

```python
# SiLU
def silu(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''
    return x/(1+np.exp(-x))

input = np.linspace(-6, 6)
output = silu(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('SiLU')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('SiLU.png')
plt.show()
```

```python
# Softplus
def softplus(x, beta):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''
    return (1/beta)* np.log(1+np.exp(beta*x))

input = np.linspace(-6, 6)
output = softplus(input,1)

plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Softplus')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Softplus.png')
plt.show()
```

```python
# Mish
def mish(x):
    return x*np.tanh(softplus(x,1))
input = np.linspace(-6, 6)
output = mish(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Mish')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Mish.png')
plt.show()
```

```python
# Softshrink
def softshrink(x,lamda):
    x1=[]
    for i in x:
        if i<-lamda:
            x1.append(i+lamda)
        elif i>lamda:
            x1.append(i-lamda)
        else:
            x1.append(0)

    return x1
input= np.linspace(-8, 8)
output = softshrink(input, 0.5)

plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Softshrink')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Softshrink.png')
plt.show()
```

```python
# Softsign
def softsing(x):
    return x/(1+ np.abs(x))
input = np.linspace(-6, 6)
output = softsing(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Softsign')
plt.xlim([-6, 6])
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Softsign.png')
plt.show()
```

```python
# Tanh
def tanh(x):
    ''' It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.'''

    return np.tanh(x)

input= np.linspace(-6, 6)
output = tanh(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Tanh')
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Tanh.png')
plt.show()
```

```python
# Tanhshrink
def tanhshrink(x):
    return x-tanh(x)
input = np.linspace(-6, 6)
output = tanhshrink(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Tanhshrink')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Tanhshrink.png')
plt.show()
```

```python
# Softmax
def softmax(x):
    ''' Compute softmax values for each sets of scores in x. '''
    return np.exp(x) / np.sum(np.exp(x), axis=0)

input= np.linspace(-6, 6)
output = softmax(input)
plt.plot(input,output, label='train')
plt.grid(True)
plt.title('Softmax')
plt.axis('tight')
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig('Softmax.png')
plt.show()
```
